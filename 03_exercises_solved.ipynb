- bcc: Cr, Fe, Mn, Ti, V

- hcp: Co, Ni, Sc, Ti, Zn

Use Python sets to group them and determine the following:

  1. Which metals only exist in one of the crystal structures?
      - fcc, 
      - bcc, 
      - hcp.

  2. Which metals exist in only two of these structures?
      - fcc+bbc,
      - fcc+hcp,
      - bcc+hcp.
  
  3. Which metals do **not** form an hcp structure?
  
  4. Do any of them exist in all three structures?

 Beware: *The union operator has lower precedence than the difference operator*, so you might need parentheses.

</div>
<div style="width: 4%;">
</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Die Kristallgitterstrukturen einiger Elemente sind unten angegeben. Einige Elemente haben unter verschiedenen Temperatur- und Druckbedingungen unterschiedliche Kristallstrukturen. 

- fcc: Cu, Co, Fe, Mn, Ni, Sc

- bcc: Cr, Fe, Mn, Ti, V

- hcp: Co, Ni, Sc, Ti, Zn

Benutzen Sie Python-Sets, um sie zu gruppieren und folgendes zu bestimmen:

  1. Welche Metalle kommen nur in einer der Kristallstrukturen vor?
      - fcc, 
      - bcc, 
      - hcp?
  
  2. Welche Metalle kommen nur in zwei dieser Strukturen vor?
      - fcc+bbc,
      - fcc+hcp,
      - bcc+hcp.
  
  3. Welche Metalle bilden **keine** hcp-Struktur?
  
  4. Kommt eines von ihnen in allen drei Strukturen vor?

Achtung: *Der Vereinigungsoperator hat einen niedrigeren Vorrang als der Differenzoperator*, so dass Sie möglicherweise Klammern benötigen.

</div>
</div>

````python
fcc = {'Cu', 'Co', 'Fe', 'Mn', 'Ni', 'Sc'}
bcc = {'Cr', 'Fe', 'Mn', 'Ti', 'V'}
hcp = {'Co', 'Ni', 'Sc', 'Ti', 'Zn'}


# fcc only but NOT bcc and hcp
print(fcc - bcc - hcp)

# fcc only but NOT bcc and hcp
print(fcc - (bcc | hcp))

# bcc only 
print(bcc - fcc - hcp)

# hcp only 
print(hcp - fcc - bcc)

# fcc AND bcc but NOT hcp
print(fcc & bcc - hcp)

# fcc AND hcp but NOT bcc
print(fcc & hcp - bcc)

# bcc AND hcp but NOT fcc
print(bcc & hcp - fcc)

# NOT hcp, but either fcc OR bcc
print((fcc | bcc) - hcp)   # the union operator has lower priority than the intersection

# in fcc AND bcc AND hcp
print(fcc & bcc & hcp)
````

#### Explanation

Consider the polynomial: 
$$ 
p(x) = 4 + 1x + 7x^2 = 4\cdot x^0 + 1\cdot x^1 + 7\cdot x^2 = \sum_i \mathrm{coeff}_i \cdot x^{\mathrm{power}_i}
$$

The data associated with this polynomial can be viewed as a set of power-coefficient pairs: in this case the coefficient 4 belongs to power 0, the coefficient 1 belongs to power 1, and the coefficient 7 belongs to power 2. 

A dictionary can be used to map a power to a coefficient:
```python
p = {0: 4, 1: 1, 2: 7}
```

A list can also be used here, with the assignment being implicit, as *each index must correspond directly to the respective power*:
```python
p = [4, 1, 7]
```

p = {0: 4, 1: 1, 2: 7}

p = [4, 1, 7]
