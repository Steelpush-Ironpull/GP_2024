````python
from IPython.display import display, clear_output, HTML
import time

def countdown_timer(minutes):
    total_seconds = minutes * 60
    for seconds in range(total_seconds, 0, -1):
        mins, secs = divmod(seconds, 60)
        time_str = f"{mins:02}'{secs:02}''"
        clear_output(wait=True)
        # HTML with styling
        display(HTML(f'<div style="font-size: 24px; color: blue; font-weight: bold;">Time remaining: {time_str}</div>'))
        time.sleep(1)
    clear_output(wait=True)
    # Final message with different styling
    display(HTML('<div style="font-size: 24px; color: green; font-weight: bold;">Time\'s up!</div>'))
````

**Seaborn** is a Python library for data visualization built on top of **Matplotlib**, the foundational library for plotting. 

If you're new to data visualization in Python, it might seem natural to begin with Matplotlib, the foundational library for plotting. However, there are some compelling reasons to start with Seaborn instead: 

1. Seaborn is Beginner-Friendly: 
Seaborn is designed with simplicity in mind. Seaborn abstracts away much of the complexity of Matplotlib, offering simple commands for creating complex visualizations with a single function call.

2. Professional-Looking Plots by Default:
Seaborn’s default styles are clean and professional. It automatically adjusts colors, grids, and other aesthetic elements to make your plots visually appealing. 

3. Seamless Integration with Pandas DataFrames: 
Most of the data you work with will come in the form of a DataFrame (e.g., using Pandas). Seaborn is built to work directly with DataFrames, allowing you to refer to columns by name instead of slicing data manually.

4. Specialized Statistical Plots:
Seaborn includes built-in tools for creating statistical plots that would require significant effort to implement in Matplotlib.

But remember: 
While *Seaborn is excellent for quickly creating beautiful and statistically informative plots, Matplotlib provides complete control* over every aspect of your plots. If you need to customize your visualizations beyond what Seaborn offers (such as adding multiple subplots, intricate annotations, or unconventional layouts) Matplotlib is indispensable.
        
</div>

## Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using *a cloud of points, where each point represents an observation* in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

## Streudiagramme

Die Punktwolke ist ein Hauptbestandteil der statistischen Visualisierung. Es stellt die gemeinsame Verteilung zweier Variablen mit Hilfe *einer Punktwolke dar, wobei jeder Punkt eine Beobachtung* im Datensatz darstellt. Diese Darstellung ermöglicht es dem Auge, eine beträchtliche Menge an Informationen darüber abzuleiten, ob es eine sinnvolle Beziehung zwischen den Variablen gibt.

</div>
</div>

````python
# Load sample data
tips = sns.load_dataset('tips')

plt.figure(figsize=(5, 4))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.title('Tips vs. Total Bill')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

- `sns.scatterplot()` is a Seaborn function for creating a scatter plot.

- `data=tips` specifies the dataset to use.

- `x='total_bill'` specifies which column in the dataset to use for the x-axis.

- `y='tip'` specifies which column in the dataset to use for the y-axis.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

- `sns.scatterplot()` ist eine Seaborn-Funktion zur Erstellung eines Streudiagramms.

- Mit `data=tips` wird der zu verwendende Datensatz angegeben.

- `x='total_bill'` gibt an, welche Spalte des Datensatzes für die x-Achse verwendet werden soll.

- `y='tip'` gibt an, welche Spalte des Datensatzes für die y-Achse zu verwenden ist.

</div>
</div>

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

- `plt.figure(figsize=(5, 4))` sets the overall figure size in inches.

- `plt.title('Tips vs. Total Bill')` adds a title to the plot; like all labels, it can be customized with f-strings.

- `plt.show()` finalizes the visualization, rendering the figure in the notebook or terminal. Without `plt.show()`, the plot might not display in some environments.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

- Mit `plt.figure(figsize=(5, 4))` wird die Gesamtgröße der Figur in Zoll festgelegt.

- `plt.title('Tips vs. Total Bill')` fügt dem Diagramm einen Titel hinzu; wie alle Beschriftungen kann er mit f-strings angepasst werden.

- `plt.show()` schließt die Visualisierung ab und stellt die Abbildung im Notebook oder Terminal dar. Ohne `plt.show()` wird die Darstellung in einigen Umgebungen möglicherweise nicht angezeigt.

</div>
</div>

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

### Semantic mapping

While the points are plotted in two dimensions, another dimension can be added to the plot by coloring the points according to a third variable. 

You can use different markers:

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

### Semantisches Mapping

Während die Punkte in zwei Dimensionen aufgetragen werden, kann dem Diagramm eine weitere Dimension hinzugefügt werden, indem die Punkte entsprechend einer dritten Variable eingefärbt werden. 

Sie können verschiedene Marker verwenden:

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', style="time")

plt.title('Tips vs. Total Bill')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

You can also use different colors.
In seaborn, this is referred to as using a *hue semantic*, because the color of the point gains meaning:

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Sie können auch verschiedene Farben verwenden.
In Seaborn wird dies als *Farbton semantisch* bezeichnet, da die Farbe des Punktes an Bedeutung gewinnt:

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue="smoker")

plt.title('Tips vs. Total Bill')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

In the previous example, the hue semantic was *categorical* (yes/no), so the default qualitative palette was applied. If the hue semantic is numeric, the default coloring switches to a *sequential* palette:

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Im vorigen Beispiel war der Farbton semantisch *kategorisch* (ja/nein), daher wurde die qualitative Standardpalette verwendet. Wenn der semantische Farbton numerisch ist, wird die Standardfarbgebung auf eine *sequenzielle* Palette umgestellt:

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue="size")

plt.title('Tips vs. Total Bill')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

[More color palettes.](https://seaborn.pydata.org/tutorial/color_palettes.html)

---

The third kind of semantic variable changes the size of each point:

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

[Weitere Farbpaletten.](https://seaborn.pydata.org/tutorial/color_palettes.html)

---

Die dritte Art von semantischer Variable verändert die Größe der einzelnen Punkte:

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', size="size")

plt.title('Tips vs. Total Bill')

plt.show()
````

### Aggregation and uncertainty

More complex datasets will have multiple measurements for the same value of the `x` variable:
</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

### Aggregation und Unsicherheit

Bei komplexeren Datensätzen gibt es mehrere Messungen für denselben Wert der Variablen `x`:

</div>
</div>

````python
fmri = sns.load_dataset("fmri")

fmri
````

````python
plt.figure(figsize=(5, 4))

# Line plot
sns.scatterplot(data=fmri, x="timepoint", y="signal")

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the *95% confidence interval* around the mean.

The 95% confidence interval quantifies the certainty of the mean.
  - Example: The average height is between 170 cm and 180 cm with 95% certainty.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Standardmäßig fasst seaborn die Mehrfachmessungen für jeden x-Wert zusammen, indem es den Mittelwert und das *95 %-Konfidenzintervall* um den Mittelwert herum aufträgt.

Das 95%-Konfidenzintervall quantifiziert die Sicherheit des Mittelwertes.
  - Beispiel: Die durchschnittliche Körpergröße liegt mit 95%iger Sicherheit zwischen 170 cm und 180 cm.

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal")

sns.scatterplot(data=fmri, x="timepoint", y="signal", alpha=0.5, marker='.')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

Another good option, especially with larger data, is to represent the spread of the distribution at each timepoint by plotting the *standard deviation* instead of a confidence interval.

The standard deviation quantifies how spread out the data is.
  - Example: Most people’s height is within 10 cm of the average.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Eine weitere gute Möglichkeit, vor allem bei größeren Datenmengen, ist die Darstellung der Streuung der Verteilung zu jedem Zeitpunkt, indem die *Standardabweichung* anstelle eines Konfidenzintervalls aufgetragen wird.

Die Standardabweichung gibt an, wie groß die Streuung der Daten ist.
  - Beispiel: Die Größe der meisten Menschen liegt innerhalb von 10 cm des Durchschnitts.

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", errorbar="sd")

sns.scatterplot(data=fmri, x="timepoint", y="signal", alpha=0.5, marker='.')

plt.show()
````

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

### Semantic mapping

The `lineplot()` function has the same flexibility as `scatterplot()`: it can show up to three additional variables by modifying the hue, size, and style of the plot elements. 

Using semantics in `lineplot()` will also determine how the data get aggregated. For example, adding a hue semantic with two levels splits the plot into two lines and error bands, coloring each to indicate which subset of the data they correspond to.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

### Semantisches Mapping

Die Funktion `lineplot()` hat die gleiche Flexibilität wie `scatterplot()`: sie kann bis zu drei zusätzliche Variablen anzeigen, indem sie den Farbton, die Größe und den Stil der Plot-Elemente verändert.

Die Verwendung von Semantiken in `lineplot()` bestimmt auch, wie die Daten aggregiert werden. Wenn man zum Beispiel eine Farbtonsemantik mit zwei Ebenen hinzufügt, wird das Diagramm in zwei Linien und Fehlerbänder aufgeteilt, die jeweils eingefärbt werden, um anzuzeigen, welcher Teilmenge der Daten sie entsprechen.

</div>
</div>

````python
plt.figure(figsize=(5, 4))

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")

plt.show()
````

## Overview of Seaborn's Plots

### 1. **Scatter Plot**

**Use Case**: Visualize relationships between two continuous variables.

```python
# Load sample data
tips = sns.load_dataset('tips')

# Scatter plot with color by category
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, style='sex', palette='Set1')
plt.title('Scatter Plot of Tips vs. Total Bill')
plt.show()
```

**Advanced Options**:
- **Add a regression line** using `sns.lmplot()` for linear regression:
    ```python
    sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
    plt.title('Scatter Plot with Regression Line')
    ```

### 2. **Line Plot**

**Use Case**: Display trends over time or sequences.

```python
# Load data for line plot
fmri = sns.load_dataset('fmri')

# Create a line plot with error bands
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri, ci='sd')
plt.title('Line Plot of FMRI Data')
plt.show()
```

### 3. **Histogram and KDE Plot**

**Use Case**: Show the distribution of a single variable.

```python
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title('Histogram of Total Bill with KDE')
plt.show()
```

**Kernel Density Estimate (KDE) Plot**:
```python
sns.kdeplot(data=tips['total_bill'], shade=True)
plt.title('KDE Plot of Total Bill')
```

### 4. **Bar Plot**

**Use Case**: Visualize categorical data comparisons.

```python
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, ci=None, palette='Blues_d')
plt.title('Total Bill by Day')
plt.show()
```

### 5. **Box Plot and Violin Plot**

**Use Case**: Display the distribution of data, highlighting medians, quartiles, and outliers.

**Box Plot**:
```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

**Violin Plot**:
```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True, palette='muted')
plt.title('Violin Plot of Total Bill by Day and Sex')
plt.show()
```

### 6. **Heatmap**

**Use Case**: Visualize matrix data with color-coded cells.

```python
# Create a correlation matrix
corr = tips.corr()

# Heatmap of the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Tips Correlation Matrix')
plt.show()
```

### 7. **Pair Plot**

**Use Case**: Explore pairwise relationships in a dataset.

```python
# Pair plot with hue for categorization
sns.pairplot(tips, hue='sex', diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Tips Data', y=1.02)
plt.show()
```

### 8. **Facet Grid**

**Use Case**: Visualize data subsets across multiple subplots.

```python
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total Bill", "Tip")
plt.show()
```

## Customization and Styling

Seaborn allows you to customize your plots extensively:

- **Change the Theme**:
    ```python
    sns.set_theme(style="whitegrid")
    ```

- **Set Color Palettes**:
    ```python
    sns.set_palette('Dark2')
    ```

- **Customize Figure Size**:
    ```python
    plt.figure(figsize=(10, 6))
    ```

- **Add Titles and Labels**:
    ```python
    plt.title('Your Plot Title')
    plt.xlabel('X-Axis Label')
    plt.ylabel('Y-Axis Label')
    ```
