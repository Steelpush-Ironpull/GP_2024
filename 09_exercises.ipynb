A quick way to visualize a regression is by using `sns.regplot()`.
Try using it in the same way you would use `sns.lineplot()`, but consider that `sns.regplot()` does not take the `hue` argument. (1 P) 

However, `regplot()` is primarily a plotting function and does not return a regression model (i.e., an equation).

Fit a linear model to `body_mass_g` vs. `flipper_length_mm` using `curve_fit` and `r2_score` as demonstrated in the lecture on NumPy. You do not need to distinguish between `species`. (1 P)

Create a plot that includes a label displaying the line equation and the $R^2$ value formatted as LaTeX. Use an f-string to format the label. (1 P)

</div>
<div style="width: 48%; line-height: 1.5;<div style="width: 48%; line-height: 1.5;color: grey;">

Eine schnelle Möglichkeit, eine Regression zu visualisieren, ist die Verwendung von `sns.regplot()`.
Versuchen Sie, diese Funktion auf die gleiche Weise zu verwenden wie `sns.lineplot()`, aber beachten Sie, dass `sns.regplot()` das Argument `hue` nicht akzeptiert. (1 P) 

Allerdings ist `regplot()` in erster Linie eine Darstellungsfunktion und gibt kein Regressionsmodell (d.h. eine Gleichung) zurück.

Passen Sie ein lineares Modell an `body_mass_g` vs. `flipper_length_mm` an, indem Sie `curve_fit` und `r2_score` verwenden, wie in der Vorlesung über NumPy gezeigt. Sie brauchen nicht zwischen `species` zu unterscheiden. (1 P)

Erstellen Sie ein Diagramm, das eine Beschriftung enthält, die die Geradengleichung und den $R^2$-Wert in LaTeX formatiert. Verwenden Sie einen f-string, um die Beschriftung zu formatieren. (1 P)
   
</div>
</div>

```python
# Line Plot
sns.regplot(data=penguins, x='flipper_length_mm', y='body_mass_g')

plt.show()

from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# Define a linear function for curve fitting
def linear_func(x, a, b):
    return a * x + b

# Extract the relevant data
x = penguins['flipper_length_mm'].dropna()
y = penguins['body_mass_g'].dropna()

# Fit the data using curve_fit
params, _ = curve_fit(linear_func, x, y)
a, b = params

r2 = r2_score(y, linear_func(x, a, b))

# Plot the data and fitted line
plt.figure(figsize=(5, 4))
sns.scatterplot(x=x, y=y, label='Data', color='skyblue')
plt.plot(x, linear_func(x, a, b), color='red', label=f'y = {a:.2f}x + {b:.2f}\nR$^2$ = {r2:.3f}')

plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend()
plt.show()
```
