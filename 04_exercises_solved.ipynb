##### Paper folding revisited

How many times must a sheet of paper with  thickness, $t = 0.1~\mathrm{mm}$ be folded to reach the moon (distance from Earth, $d = 384400~\mathrm{km}$)?

**Mathematical Solution** (already solved in class)

Each time we fold the paper, we double its thickness. Therefore, if we fold the paper $n$ times, its thickness becomes $t \cdot 2^n$. 
Setting $t \cdot 2^n \geq d$ gives us $n = \log_2\left(\frac{d}{t}\right)$.

</div>

<div style="width: 48%; line-height: 1.5;color: grey;">

##### Papierfalten neu interpretiert

Wie oft muss ein Blatt Papier mit der Dicke $t = 0.1~\mathrm{mm}$ gefaltet werden, um den Mond zu erreichen (Entfernung von der Erde, $d = 384400~\mathrm{km}$)?

**Mathematische Lösung** (bereits im Unterricht gelöst)

Jedes Mal, wenn wir das Papier falten, verdoppeln wir seine Dicke. Wenn wir also das Papier $n$ mal falten, wird es $t \cdot 2^n$ dick.
Setzt man $t \cdot 2^n \geq d$, so erhält man $n = \log_2\left(\frac{d}{t}\right)$.

</div>
</div>

$$
t \cdot 2^n \geq d \implies n = \log_2\left(\frac{d}{t}\right)
$$

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

**Solving with a `while` loop**

To avoid math and solve this with a `while` loop, follow these steps:

- Start with a sheet of paper with an initial thickness of 0.1 millimeters.

- Set a target distance to the moon (384400 km).

- Use a `while` loop to keep doubling the thickness with each fold until it reaches or exceeds the distance to the moon.

- Count each fold as the loop iterates.

<details>
  <summary>Hint</summary>

Recall from the lecture on variables (`01_python_basic_calculator.ipynb`): to double the value of a variable in place, use the command `t *= 2`. This is known as an augmented assignment and is shorthand for `t = t * 2`.
</details>

</div>

<div style="width: 48%; line-height: 1.5;color: grey;">

**Lösen mit einer `while`-Schleife**

Um Mathematik zu vermeiden und dieses Problem mit einer `while`-Schleife zu lösen, gehen Sie folgendermaßen vor:

- Beginnen Sie mit einem Blatt Papier mit einer Anfangsdicke von 0.1 Millimetern.

- Legen Sie eine Zielentfernung zum Mond fest (384400 km).

- Verwenden Sie eine `while`-Schleife, um die Dicke mit jeder Faltung zu verdoppeln, bis sie die Entfernung zum Mond erreicht oder überschreitet.

- Zählen Sie jede Falte, während die Schleife iteriert.

<details>
  <summary>Hinweis</summary>
  
Erinnern Sie sich an die Vorlesung über Variablen (`01_python_basic_calculator.ipynb`): Um den Wert einer Variablen an Ort und Stelle zu verdoppeln, verwenden Sie den Befehl `t *= 2`. Dies wird als erweiterte Zuweisung bezeichnet und ist die Kurzform für `t = t * 2`.
</details>

</div>
</div>

````python
thickness = 0.1  # in millimeters
distance_to_moon = 384_400 * 1_000_000  # convert kilometers to millimeters

# Initialize variables
folds = 0

# While loop to double the thickness with each fold
while thickness < distance_to_moon:
    thickness *= 2
    folds += 1

# Result
print(f"It takes {folds} folds for the paper to reach the distance to the moon.")
````

Die Fakultät einer Zahl (bezeichnet als `n!`) ist das Produkt aller positiven ganzen Zahlen von n bis hinunter zu 1. 

Führen Sie die folgenden Schritte aus:

- Initialisiere eine Variable `factorial` mit einem Wert von 1, um das Ergebnis zu speichern.

- Verwenden Sie eine `for`-Schleife und die Funktion `range()`, um durch die Zahlen von 1 bis n (wählen Sie `n` und beachten Sie dabei, dass `5` enthalten sein sollte).

The Fibonacci sequence is a series of numbers in which the next number is found by adding up the two numbers before it. 

The first two numbers are $0$ and $1$. 

For example: $0, 1, 1, 2, 3, 5, 8, 13, 21$. 

The next number in this series is $13 + 21 = 34$.
</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Die Fibonacci-Folge ist eine Zahlenreihe, bei der die nächste Zahl durch die Addition der beiden vorhergehenden Zahlen gefunden wird. 

Die ersten beiden Zahlen sind $0$ und $1$. 

Zum Beispiel: $0, 1, 1, 2, 3, 5, 8, 13, 21$. 

Die nächste Zahl in dieser Reihe ist $13 + 21 = 34$.
</div>
</div>
<div style="text-align: center; margin-top: 20px;">

$$
0 \xrightarrow{+1} 1 \xrightarrow{+0} 1 \xrightarrow{+1} 2 \xrightarrow{+1} 3 \xrightarrow{+2} 5 \xrightarrow{+3} 8 \xrightarrow{+5} 13 \xrightarrow{+8} 21 \xrightarrow{+13} 34
$$

</div>

<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

A Collatz sequence is a series of numbers generated in the following way:

 Start with any positive integer;
 If this number is even, divide by two to generate the next number in the series;
 If it is instead odd, multiply by three and add one;
 Repeat this process to generate subsequent members of the sequence.


The Collatz conjecture (which remains unproven) suggests that no matter the starting number, the sequence will eventually reach 1. However, the length of the sequence varies depending on the initial number. For instance, starting with 3 produces a Collatz sequence with a length of 8:

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Eine Collatz-Folge ist eine Reihe von Zahlen, die auf folgende Weise erzeugt werden:

 Beginne mit einer beliebigen positiven ganzen Zahl;
 Wenn diese Zahl gerade ist, dividiere durch zwei, um die nächste Zahl in der Reihe zu erzeugen;
 Ist sie stattdessen ungerade, multipliziere mit drei und addiere eins;
 Wiederhole diesen Vorgang, um die nächsten Glieder der Reihe zu erzeugen.

Die Collatz-Vermutung (die nach wie vor unbewiesen ist) besagt, dass die Folge unabhängig von der Ausgangszahl schließlich 1 erreichen wird. Die Länge der Folge variiert jedoch in Abhängigkeit von der Ausgangszahl. Wenn man beispielsweise mit 3 beginnt, ergibt sich eine Collatz-Folge mit einer Länge von 8:
</div>
</div>
<div style="text-align: center; margin-top: 20px;">

$$
3\xrightarrow{\times 3,\, +1} 10\xrightarrow{\div2}5\xrightarrow{\times 3,\, +1}16\xrightarrow{\div2}8\xrightarrow{\div2}4\xrightarrow{\div2}2\xrightarrow{\div2}1
$$
</div>
<div style="display: flex; justify-content: space-between;">
<div style="width: 48%; line-height: 1.5;">

Use a `while` loop to determine the length of the Collatz sequence starting with 837799, which holds the record for the longest Collatz sequence of any number below 1,000,000.
</div>
<div style="width: 48%; line-height: 1.5;color: grey;">

Verwenden Sie eine `while`-Schleife, um die Länge der Collatz-Folge zu bestimmen, die mit 837799 beginnt. Diese Zahl hält den Rekord für die längste Collatz-Folge einer Zahl unter 1.000.000.
</div>
</div>

````python
n = 837799

length = 1              # start measuring the length
while n != 1:           # as long as n is not 1
    if n % 2 == 0:      # if n is even, divide it by 2
        n = n / 2
    else:
        n = 3 * n + 1   # if n is odd, multiply it by 3 and add 1
    length += 1         # update the length
print(length)
````

A prime number is a number that cannot be made by multiplying other whole numbers. 

A prime number is a natural number greater than 1 that is not a product of two smaller natural numbers. 

Print all prime numbers between 25 and 50.

*Hint*

Use a nested loop to check each number in the given range. If a number is found to have a divisor other than 1 and itself, use the `break` statement to stop checking further divisors and move on to the next number. The `break` statement helps you skip numbers that are confirmed to be non-prime.

</div>
<div style="width: 48%; line-height: 1.5;color: grey;">


Eine Primzahl ist eine Zahl, die nicht durch Multiplikation anderer ganzer Zahlen gebildet werden kann. 

Eine Primzahl ist eine natürliche Zahl größer als 1, die kein Produkt aus zwei kleineren natürlichen Zahlen ist. 

Drucke alle Primzahlen zwischen 25 und 50 aus.

*Hinweis*

Verwenden Sie eine verschachtelte Schleife, um jede Zahl im angegebenen Bereich zu überprüfen. Wird festgestellt, dass eine Zahl einen anderen Teiler als 1 und sich selbst hat, verwenden Sie die Anweisung `break`, um die Prüfung weiterer Teiler zu beenden und zur nächsten Zahl überzugehen. Die Anweisung `break` hilft Ihnen, Zahlen zu überspringen, die sich als nicht-primzahlig herausstellen.

</div>
</div>

````python
primes = []

start = 25
end = 50

print("Prime numbers between", start, "and", end, "are:")

for num in range(start, end + 1):
    for i in range(2, num):  # Check for factors
        if (num % i) == 0:   # Not a prime number, break the inner loop and look for the next number
            break
    else:
        primes.append(num)

print(primes)
````
