````python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Inspect the dataset
print("First few rows of the Titanic dataset:")
print(titanic.head(), "\n")

print("Basic information about the dataset:")
print(titanic.info(), "\n")

print("Shape of the dataset:")
print(titanic.shape, "\n")

# Create a reduced dataset
columns_to_keep = ['survived', 'pclass', 'sex', 'age', 'fare', 'deck', 'embark_town', 'alone']
titanic_clean = titanic[columns_to_keep]
print("Reduced dataset preview:")
print(titanic_clean.head(), "\n")

# Handle missing data
titanic_clean = titanic_clean.dropna(subset=['age'])
print("Dataset after dropping rows with missing 'age' values:")
print(titanic_clean.info(), "\n")

# General Statistics
total_passengers = len(titanic_clean)
total_males = len(titanic_clean[titanic_clean['sex'] == 'male'])
total_females = len(titanic_clean[titanic_clean['sex'] == 'female'])
total_children = len(titanic_clean[titanic_clean['age'] < 14])
total_alone = len(titanic_clean[titanic_clean['alone'] == True])

print("General Statistics:")
print(f"  Total passengers: {total_passengers}")
print(f"  Males: {total_males}")
print(f"  Females: {total_females}")
print(f"  Children (<14 years): {total_children}")
print(f"  Passengers traveling alone: {total_alone}\n")

# Pclass-specific Survival Statistics
print("Pclass-specific Survival Statistics:")
survivors = titanic_clean[titanic_clean['survived'] == 1]
for pclass in sorted(survivors['pclass'].unique()):
    class_survivors = len(survivors[survivors['pclass'] == pclass])
    total_class_passengers = len(titanic_clean[titanic_clean['pclass'] == pclass])
    class_survival_rate = (class_survivors / total_class_passengers) * 100 if total_class_passengers > 0 else 0

    print(f"  Pclass {pclass}:")
    print(f"    Survivors: {class_survivors}")
    print(f"    Survival rate: {class_survival_rate:.2f}%")
````
